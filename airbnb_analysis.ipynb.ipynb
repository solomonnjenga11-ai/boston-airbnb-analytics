```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('Listings.csv')

# Define all columns to drop
columns_to_drop = [
    'source', 'name', 'description', 'neighborhood_overview', 'picture_url',
    'host_id', 'host_url', 'host_name', 'host_about', 'host_thumbnail_url',
    'host_picture_url', 'host_verifications', 'neighbourhood', 'amenities',
    'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights',
    'maximum_maximum_nights', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm',
    'calendar_updated', 'license', 'id', 'listing_url', 'scrape_id',
    'reviews_per_month'
]

# Drop the columns safely
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore')

# Confirm new shape
print(f"Cleaned dataset: {df_cleaned.shape[0]} rows, {df_cleaned.shape[1]} columns")

# Preview cleaned data
df_cleaned.head()
```

```python
# Full transformation code


# Step 1: Copy the original cleaned dataset
df_transformed = df_cleaned.copy()

# Step 2: Convert 't' and 'f' to 1 and 0 in all applicable object columns
binary_columns = df_transformed.select_dtypes(include='object').columns
for col in binary_columns:
    df_transformed[col] = df_transformed[col].replace({'t': 1, 'f': 0})

# Step 3: Replace 'neighbourhood_cleansed' with ordinal values based on average price_per_person
neighborhood_avg = df_transformed.groupby('neighbourhood_cleansed')['price_per_person'].mean().sort_values()
neighborhood_map = {name: idx for idx, name in enumerate(neighborhood_avg.index)}
df_transformed['neighbourhood_cleansed'] = df_transformed['neighbourhood_cleansed'].map(neighborhood_map)

# Step 4: Compute host_tenure_years from host_since and last_scraped
df_transformed['host_since'] = pd.to_datetime(df_transformed['host_since'], errors='coerce')
df_transformed['last_scraped'] = pd.to_datetime(df_transformed['last_scraped'], errors='coerce')
df_transformed['host_tenure_years'] = (
    (df_transformed['last_scraped'] - df_transformed['host_since']).dt.total_seconds() / (365.25 * 24 * 3600)
)

# Step 5: Convert host_response_time to ordinal values
response_order = {
    'within an hour': 0,
    'within a few hours': 1,
    'within a day': 2,
    'a few days or more': 3
}
df_transformed['host_response_time'] = df_transformed['host_response_time'].map(response_order)

# Step 6: Drop unwanted columns
df_transformed.drop(columns=['bathrooms_text', 'host_location'], errors='ignore', inplace=True)

# Step 7: Preview the transformed dataset
print(f"Transformed dataset shape: {df_transformed.shape}")
df_transformed.head()



```
